```python
url = "https://sports.news.naver.com/kbaseball/record/index?category=kbo&year=" #url
year = "2022" #year
years = ["2016", "2017", "2018", "2019", "2020", "2021", "2022"] # 리스트에서 원소를 가져오는 인덱싱
```

```python
df = pd.DataFrame([]) #빈프레임 설정
for i in years:      # for문 생성
    df1 = pd.read_html(url + i)[0] # df1 지정
    df = pd.concat([df, df1]) # df 프레임 쌓임
df # df 실행
```

```python
df = pd.DataFrame([]) #빈프레임 설정
for i in years:      # for문 생성
    df1 = pd.read_html(url + i)[0] # df1 지정
    df1["연도"] = i # 연도 열 생성
    df = pd.concat([df, df1]) # df 프레임 쌓임
df # df 실행
```

```python
url = "https://sports.news.naver.com/kbaseball/record/index?category=kbo&year=" #url
year = "2022" #year
years = range(2015, 2023) # 리스트에서 원소를 가져오는 인덱싱 #range 함수
```

```python
df = pd.DataFrame([]) #빈프레임 설정
for i in years:      # for문 생성
    df1 = pd.read_html(url + str(i))[0] # df1 지정 #string 문자열 리스트
    df1["연도"] = (i) # 연도 열 생성
    df = pd.concat([df, df1]) # df 프레임 쌓임
df # df 실행
```

```python
# 2번의 결과물을 다양한 방법으로 피벗해본다.
df.pivot(values="순위", index="연도", columns="팀")
```
