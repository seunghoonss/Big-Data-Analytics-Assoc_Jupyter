```python
import pandas as pd
a1 = [60, 84, 80, 70, 19]
a2 = [77, 62, 95, 85, 17]
a3 = [61, 97, 72, 67, 15]
a4 = [75, 65, 95, 51, 18]
cols = ["국어", "영어", "수학", "과학", "나이"]
df = pd.DataFrame([a1, a2, a3, a4], index=list("ABCD"), columns=cols)
df
```

```python
df.sort_values("국어") # 오름차순 디폴트
```

```python
df.sort_values("영어", ascending=False) # 내림차순
```

```python
df.sort_values("수학", ascending=False) # 동점자에서는 먼저 위에 있던 행이 위로
```

```python
df.sort_values(["수학", "영어"], ascending=False)
```

```python
df.sort_values(["수학", "나이"], ascending=[0, 1]) # False = 0, True =1
```
